```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('/content/drive/MyDrive/data_teste/Salary_Data.csv', delimiter=',')
df.info
```

```python
df.head(5)
```

# Alterando o título dos campos

```python
df.rename(columns={
    'Age':'Idade',
    'Gender':'Gênero',
    'Education Level':'Nível de Educação',
    'Job Title':'Título Emprego',
    'Years of Experience':'Experiência',
    'Salary':'Salário'
},inplace=True)

df.head(5)
```

# Verificando Campos nulos

```python
plt.figure(figsize=(12,4))
plt.title("Verificando campos nulos")
sns.heatmap(df.isnull());
```

#Avaliar salário por Gênero:

```python
plt.figure(figsize=(10,5))
plt.title("Análise salário x Gênero")
plt.style.use('ggplot')

sns.barplot(data=df, x='Gênero', y='Salário',errorbar=None );
```

#Análise de salário por formação

```python
plt.figure(figsize=(14,5))
plt.title("Análise salário x Nível de educação")

sns.boxplot(data=df, x="Nível de Educação", y="Salário", palette='pastel');
```

```python
plt.figure(figsize=(12,5))
plt.title('Análise de experiência por idade')
sns.lineplot(data=df, x='Experiência', y='Idade');
```

#Análise de idade por gênero:

```python
plt.figure(figsize=(10,4))
plt.title('Análise de idade por gênero')
sns.boxplot(data=df, x='Gênero', y='Idade', palette='pastel');
```
